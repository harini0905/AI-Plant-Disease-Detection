#  🤖 Hugging Face Generative AI Integration

This notebook demonstrates the integration of Hugging Face Generative AI with the Plant Disease Detection System.

After the deep learning model predicts a plant disease, the detected disease name is passed to a Large Language Model (LLM) to generate:

- Disease Overview
- Symptoms
- Causes
- Recommended Treatment
- Prevention Tips

The generated response is designed to provide simple and farmer-friendly recommendations.

## 🔐 Loading Hugging Face API Token

The Hugging Face API token is securely loaded using a `.env` file.

Using environment variables prevents exposing sensitive API keys in the source code.

Steps:

1. Store the token in `.env` file.
2. Load the environment variables.
3. Retrieve the Hugging Face token.

```python
from dotenv import load_dotenv
import os
from huggingface_hub import InferenceClient
```

## 👤 Checking User Authentication

This step verifies the authenticated Hugging Face user details.

Successful authentication confirms that the API token is correctly configured.

```python
load_dotenv()

HF_TOKEN = os.getenv("HF_TOKEN")

print("Token Loaded:", HF_TOKEN is not None)
```

## ✅ Verifying Hugging Face Authentication

The Hugging Face account authentication is verified using the API token.

The `whoami()` function confirms whether the provided token is valid and connected to the Hugging Face account.

```python
from huggingface_hub import whoami
```

## 🔗 Initializing Hugging Face Inference Client

The Hugging Face Inference Client is initialized using the authenticated API token.

This client is used to communicate with hosted Large Language Models.

```python
from huggingface_hub import InferenceClient

client = InferenceClient(
    api_key=HF_TOKEN
)
```

## 📝 Creating Disease Recommendation Prompt

The predicted disease name from the plant disease classification model is provided as input.

A structured prompt is created to generate:

- Disease Overview
- Symptoms
- Causes
- Treatment Recommendations
- Prevention Tips

Example disease:

Tomato Early Blight

```python
predicted_disease = "Tomato Early Blight"

prompt = f"""
Disease: {predicted_disease}

Provide:

1. Disease Overview
2. Symptoms
3. Causes
4. Recommended Treatment
5. Prevention Tips

Keep the explanation simple and farmer-friendly.
"""
```

## 🤖 Generating AI-Based Treatment Recommendation

The Hugging Face Large Language Model generates a detailed response based on the detected disease.

Model Used:

**Qwen/Qwen2.5-7B-Instruct**

The generated output provides useful agricultural guidance for disease management.

```python
try:

    response = client.chat.completions.create(
        model="Qwen/Qwen2.5-7B-Instruct",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_tokens=500
    )

    print("========== AI RESPONSE ==========\n")
    print(response.choices[0].message.content)


except Exception as e:

    print("Hugging Face Error:")
    print(e)
```
